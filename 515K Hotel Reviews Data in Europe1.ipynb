```python
plt.figure(figsize=(20, 200))  # Appropriate size for horizontal bar plot
sns.barplot(y='Hotel_Name', x='Average_Score', data=ho)
plt.tight_layout()
plt.show()
```

```python
# Bar Plot was getting so much conjusted hence taken 10000 rows only two show properly
head = ho.head(10000)
```

```python
# Bar Plot
# Distribution of average scores across top 14 hotels

plt.figure(figsize=(12, 6))
sns.barplot(x='Hotel_Name', y='Average_Score', data=head)
plt.xticks(rotation=90)
plt.show()
```

```python
# Heatmap
# Number of reviews by nationality

nationality_reviews = pd.crosstab(head['Hotel_Name'], head['Reviewer_Nationality'])
plt.figure(figsize=(20, 20))
sns.heatmap(nationality_reviews, linewidths=.5)
plt.show()
```

```python
# Heatmap
# Number of reviews by nationality

nationality_reviews = pd.crosstab(head['Hotel_Name'], head['Reviewer_Nationality'])
plt.figure(figsize=(30, 200))
sns.heatmap(nationality_reviews, linewidths=.5)
plt.show()
```

```python
# Correlation Heatmap
# Correlation between numerical features

# Select only numerical columns
numerical_columns = head.select_dtypes(include=['number'])

plt.figure(figsize=(10, 6))
sns.heatmap(numerical_columns.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
# Scatterplot
# Relationship between reviewer score and word count of reviews

plt.figure(figsize=(10, 10))
sns.scatterplot(x='Review_Total_Positive_Word_Counts', y='Reviewer_Score', data=head)
plt.show()
```

```python
# Box Plot
# Distribution of reviewer scores across different hotels

plt.figure(figsize=(12, 6))
sns.boxplot(x='Hotel_Name', y='Reviewer_Score', data=head)
plt.xticks(rotation=90)
plt.show()
```
